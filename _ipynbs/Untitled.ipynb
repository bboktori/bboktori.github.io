```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

```

```python
x = np.arange(-5, 5.1, 0.1)
y = norm.pdf(x=x, loc=0, scale=1)
plt.plot(x, y)
```

```python
from scipy.stats import binom

x = np.arange(0, 6)
y = binom.pmf(x, 6, 0.5)

plt.plot(x, y, 'o')
plt.vlines(x, 0, y)
```
